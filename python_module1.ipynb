```python
#Arithmetic Operators +,-,*,/,%,//,**
#% is the remainder operator, // is integer type division and / is float type division.
a=2
b=3
print(a**b)
```

```python
#Operators with higher precedence are executed first.
#If operators have the same precedence, evaluation  usually  go from left-to-right.
#Parentheses () have the higher  precedence.
```

```python
Python follows the order of operations, also known as PEMDAS:
P: Parentheses

E: Exponents

MD: Multiplication and Division (from left to right)

AS: Addition and Subtraction (from left to right)
```

```python
#ORDER: ()  →  **  →  * , / , //, %  →  + ,-  →  < ,<= ,>, >=, ==, !=  
#→  not  →  and  →  or  →  =
```

```python
print(2**3+(2+2)**(2+1))   #72
Step-by-step calculation:
(2+2) = 4
(2+1) = 3
2**3 = 8
4**3 = 64

8 + 64 = 72
```

```python
  #floor division 
#For positive numbers, it's like regular division but discards the decimal part.
#For negative numbers, it rounds down to the next lower integer (not toward zero).

print(7 / 2)     # Output: 3      (because 7 / 2 = 3.5 → floor = 3)
```

```python
print(-7 // 2)    # Output: -4     (because -7 / 2 = -3.5 → floor = -4)
```

# Membership operators

```python
#Check if a value exists in a sequence.
# in	True if value(substring) exists	'a' in 'apple' → True
# not in	True if value missing	'z' not in 'apple' → True
```

```python
text = "Hello, world!"
print("w" in text)
```

```python
print("w" not in text)
```

```python
# Check if a substring is present
print("Hello" in text)      # True
print("world" in text)      # True
print("Python" in text)     # False

# Check if a substring is NOT present
print("Python" not in text) # True
```
