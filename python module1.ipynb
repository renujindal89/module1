Can contain letters, digits (0–9), and underscores
Python keyword (like if, for, class, etc.) CAN NOT BE USED AS IDENTIFIER

```python

  #floor division 
#For positive numbers, it's like regular division but discards the decimal part.
#For negative numbers, it rounds down to the next lower integer (not toward zero).

print(7 // 2)     # Output: 3      (because 7 / 2 = 3.5 → floor = 3)
print(-7 // 2)    # Output: -4     (because -7 / 2 = -3.5 → floor = -4)
print(7.5 // 2)   # Output: 3.0    (still works with floats)
print(10 // 3)    # Output: 3
```

Membership operators

```python
#Check if a value exists in a sequence.
# in	True if value(substring) exists	'a' in 'apple' → True
# not in	True if value missing	'z' not in 'apple' → True
```

```python
text = "Hello, world!"
print("Hello" in text)
print("renu" not in text)
```

```python
# Check if a substring is present
print("Hello" in text)      # True
print("world" in text)      # True
print("Python" in text)     # False

# Check if a substring is NOT present
print("Python" not in text) # True
```

identity operator

```python
Python tries to save memory and increase speed by reusing objects wherever it safely can.
This behavior is called "interning".

Small integers: from -5 to 256

Some strings: like short alphanumeric strings without spaces

Booleans: True and False

None (there's only one None object)
```

```python
a=1
b=1   #is operator checks if a and b refer to the exact same object in memory
print(a is b)  #interning range (-5 to 256),
```

```python
id(a)
id(b)
```

```python
a is b → compares object identity (same memory address?)
a == b → compares value equality
a=b -> both a and b refer to the same object in memory
```
